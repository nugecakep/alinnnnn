NUGE

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
img_path = "nuge.jpeg"
img = cv2.imread(img_path)
print(img.shape)
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
R, G, B = fix_img[:,:,0], fix_img[:,:,1], fix_img[:,:,2]
print(np.array(fix_img))
```

METODE LIGHTNESS NUGE

```python
#metode lightness
fix_img[:] = np.max(fix_img, axis = -1, keepdims=1)/2 + np.min(fix_img, axis = -1, keepdims=1)/2
print(np.array(fix_img[:]))
plt.axis("off")
plt.imshow(fix_img[:])
plt.savefig("Metode Lightness NUGE", bbox_inches="tight")
```

METODE AVERAGE NUGE

```python
gray_img = np.mean(fix_img, axis = -1)
print(np.array(gray_img))
plt.axis("off")
plt.imshow(gray_img, cmap = "gray")
plt.savefig("Metode Average NUGE", bbox_inches="tight")
```

Metode Luminosity/Weighted Average NUGE

```python
lumi_img = (0.2126*R) + (0.7152*G) + (0.0722*B)
print(lumi_img)
print(np.array(lumi_img))
plt.axis("off")
plt.imshow(lumi_img, cmap ="gray")
plt.savefig("Metode Luminosity NUGE", bbox_inches="tight")
```

Metode Weighted Average/Luminosity

```python
wav_img = (0.299*R) + (0.587*G) + (0.114*B)
# print(lumi_img)
print(np.array(wav_img))
plt.axis("off")
plt.imshow(wav_img, cmap = "gray")
plt.savefig("Metode Weighted Average NUGE", bbox_inches="tight")
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def convert_to_grayscale(image):
    # Menggunakan formula konversi warna ke citra grayscale
    grayscale_image = 0.299 * image[:,:,2] + 0.587 * image[:,:,1] + 0.114 * image[:,:,0]
    return grayscale_image

# Membaca citra
nuge = cv2.imread('nuge.jpeg')
jian = cv2.imread('jian.jpeg')
aril = cv2.imread('aril.jpeg')

# Memisahkan saluran warna
B_nuge, G_nuge, R_nuge = cv2.split(nuge)
B_jian, G_jian, R_jian = cv2.split(jian)
B_aril, G_aril, R_aril = cv2.split(aril)

# Menghitung citra grayscale dari setiap citra
gray_nuge = convert_to_grayscale(nuge)
gray_jian = convert_to_grayscale(jian)
gray_aril = convert_to_grayscale(aril)

# Menampilkan citra asli dan citra grayscale
plt.figure(figsize=(12, 8))

plt.subplot(3, 4, 1), plt.imshow(B_nuge, cmap='Blues'), plt.title('Blue Nuge')
plt.subplot(3, 4, 2), plt.imshow(G_nuge, cmap='Greens'), plt.title('Green Nuge')
plt.subplot(3, 4, 3), plt.imshow(R_nuge, cmap='Reds'), plt.title('Red Nuge')
plt.subplot(3, 4, 4), plt.imshow(nuge), plt.title('Original Nuge')

plt.subplot(3, 4, 5), plt.imshow(B_jian, cmap='Blues'), plt.title('Blue Jian')
plt.subplot(3, 4, 6), plt.imshow(G_jian, cmap='Greens'), plt.title('Green Jian')
plt.subplot(3, 4, 7), plt.imshow(R_jian, cmap='Reds'), plt.title('Red Jian')
plt.subplot(3, 4, 8), plt.imshow(jian), plt.title('Original Jian')

plt.subplot(3, 4, 9), plt.imshow(B_aril, cmap='Blues'), plt.title('Blue Aril')
plt.subplot(3, 4, 10), plt.imshow(G_aril, cmap='Greens'), plt.title('Green Aril')
plt.subplot(3, 4, 11), plt.imshow(R_aril, cmap='Reds'), plt.title('Red Aril')
plt.subplot(3, 4, 12), plt.imshow(aril), plt.title('Original Aril')

plt.show()
```
